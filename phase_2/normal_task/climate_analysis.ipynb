```python
# import libraries 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
daily_data=pd.read_csv('https://raw.githubusercontent.com/Aya123-sys/CVIP-data_science/main/phase_2/normal_task/daily_data.csv',parse_dates=['DATE'],na_values=[ 'T'])
```

```python
daily_data.describe()
```

```python
daily_data['Month'] = daily_data['DATE'].dt.month
```

```python
monthly_insights = daily_data.groupby('Month').agg({'DailyAverageDryBulbTemperature': 'mean', 'DailyPrecipitation': 'sum'})
print(monthly_insights)
```

```python
peak_temperature_days = daily_data[daily_data['DailyMaximumDryBulbTemperature'] > 90]
peak_temperature_days[['DATE', 'DailyMaximumDryBulbTemperature']]
```

```python
monthly_insights = daily_data.groupby('Month').agg({'DailyMaximumDryBulbTemperature': 'max'})
print(monthly_insights)
```

```python
monthly_insights[monthly_insights['DailyMaximumDryBulbTemperature']>90]
```

```python
# Visualize distribution of wind directions
plt.figure(figsize=(12, 6))
plt.hist(daily_data['DailyPeakWindDirection'].dropna(), bins=360, range=(0, 360), density=True, alpha=0.7)
plt.title('Distribution of Daily Peak Wind Directions')
plt.xlabel('Wind Direction (degrees)')
plt.ylabel('Density')
plt.show()
```

```python
monthly_insights = daily_data.groupby('Month').agg({'DailyAverageDryBulbTemperature': 'mean'})
plt.figure(figsize=(12, 6))
plt.plot(monthly_insights.index, monthly_insights.values, marker='o')
plt.title('Monthly Average Temperature')
plt.xlabel('Month')
plt.ylabel('Average Temperature')
plt.show()
```

```python
# Visualize the distribution of daily average wind speed
plt.figure(figsize=(12, 6))
plt.hist(daily_data['DailyAverageWindSpeed'].dropna(), bins=30, edgecolor='black')
plt.title('Daily Average Wind Speed')
plt.xlabel('Wind Speed')
plt.ylabel('Frequency')
plt.show()
```

```python
daily_data['Season'] = (daily_data['DATE'].dt.month % 12 + 3) // 3
seasonal_temperature = daily_data.groupby('Season')['DailyAverageDryBulbTemperature'].mean()
plt.figure(figsize=(12, 6))
plt.plot(seasonal_temperature.index, seasonal_temperature.values, marker='o')
plt.title('Seasonal Average Temperature Trends')
plt.xlabel('Season')
plt.ylabel('Average Temperature (°F)')
plt.xticks([1, 2, 3, 4], ['Winter', 'Spring', 'Summer', 'Fall'])
plt.show()
```
